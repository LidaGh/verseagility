```python
_data = data[1:50]
pd.read_json(json.dumps(_data))
```

```python
data_norm = pd.io.json.json_normalize(data, sep='_').to_dict(orient='records')
```

```python
df['cat'] = df.appliesTo.apply(lambda x: x.split(',')[-1])
df['subcat'] = df.appliesTo.apply(lambda x: x.split(',')[0])
```
